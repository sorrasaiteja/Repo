#We will now try to verify the results from simulation

```python
Cups = []
for i in range(100000):
  N_cust = np.random.poisson(500)
  totalpurchase = 0
  for j in range(N_cust):
    rpurch = np.random.uniform(0,1)
    if rpurch<0.1:
      totalpurchase = totalpurchase + 0
    elif rpurch<0.9:
      totalpurchase = totalpurchase + 1
    else:
      totalpurchase = totalpurchase + 2
  Cups.append(totalpurchase)  
```

```python
Cups
```

```python
average = sum(Cups)/len(Cups)
average
```

```python
var = sum((x-average)**2 for x in Cups) / len(Cups)
var
```

```python
varest = var*(len(Cups)/(len(Cups)-1))
varest
```
